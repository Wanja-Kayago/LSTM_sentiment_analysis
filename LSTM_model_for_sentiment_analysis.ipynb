# IMDB Movie Review Dataset

The IMDB Movie Review Dataset is one of the most popular datasets for sentiment analysis tasks, widely used in natural language processing (NLP) research and real-world applications. This dataset contains 50,000 movie reviews collected from the Internet Movie Database (IMDB), where each review is labeled as either positive or negative. It's a go-to resource for anyone looking to build, test, or improve sentiment analysis models.

## Balanced Dataset for Sentiment Analysis

One of the most important features of the IMDB dataset is its balance between positive and negative reviews. With 25,000 positive reviews and 25,000 negative reviews, the dataset ensures that no particular sentiment is over-represented, which is crucial for avoiding biased model training. This balance makes it an excellent tool for experimenting with different algorithms, as it provides a consistent benchmark for model evaluation.

## Data Structure

Columns:
review: The raw text of the movie review, which can vary in length from a few sentences to several paragraphs. This variability is useful for testing models on both short and long text inputs.
sentiment: A binary label that indicates whether the review is positive or negative. This serves as the target variable for classification tasks.

## Removal of Named Entities

```python
# After running the installation commands above, you can load the SpaCy model and use it
import spacy

# Load the SpaCy model for named entity recognition
nlp = spacy.load('en_core_web_lg')

def remove_named_entities(text):
    # Process the text with SpaCy
    doc = nlp(text)
    # Keep tokens that are not named entities
    tokens = [token.text for token in doc if not token.ent_type_]
    return ' '.join(tokens)

# Apply the named entity removal function to the dataframe
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_named_entities)
```

## Removing stop words

```python
custom_stop_words = [
    # Pronouns
    'i', 'me', 'my', 'we', 'our', 'you', 'he', 'him', 'she', 'it', 'they', 
    'this', 'that', 'these', 'those', 'everyone', 'someone',

    # Auxiliary verbs and other common verbs
    'am', 'is', 'are', 'was', 'were', 'be', 'been', 'having', 
    'can', 'could', 'should', 'would', 'may', 'must', 

    # Common prepositions and conjunctions
    'and', 'but', 'or', 'for', 'so', 'yet', 'because', 
    'with', 'about', 'between', 'during', 'before', 'after', 'without',

    # Articles and determiners
    'a', 'an', 'the', 'some', 'any', 'all', 'each', 'either', 'every', 'both',

    # Filler words and expressions
    'lot', 'things', 'stuff', 'something', 'anything', 'nothing', 
    'basically', 'really', 'anyway', 'like', 'kind', 'sort', 'type', 'seen',

    # Movie-related terms (neutral in sentiment)
    'movie', 'film', 'show', 'scene', 'scenes', 'character', 'characters', 
    'story', 'stories', 'plot', 'series', 'episode', 'episodes',
    'director', 'actors', 'acting', 'role', 'cast', 'performance', 'performances',
    'screen', 'screens', 'cinema', 'production', 'script', 'storyline', 
    'soundtrack', 'music', 'score', 'sound', 'play', 'action', 'history', 'comedy', 

    # Common phrases that don't add meaning
    'in my opinion', 'i think', 'in conclusion', 'as a whole',
    'on the other hand', 'overall', 'at times', 'in general',

    # Time references
    'hour', 'minute', 'second', 'day', 'night', 'beginning', 'middle', 'end', 'time',

    # Other common neutral words
    'thing', 'anything', 'everything', 'nothing', 'whatever', 'however', 'actually', 
    'mostly', 'seems', 'quite', 'probably', 'basically', 'especially', 'usually',
]

def remove_custom_stopwords(text):
    # Tokenize the text (split into words)
    tokens = text.split()
    # Keep tokens that are not in the custom stop words list
    tokens = [word for word in tokens if word.lower() not in custom_stop_words]
    # Return the filtered text
    return ' '.join(tokens)

# Apply the stop word removal function to the 'review' column
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_custom_stopwords)
```

## Removing filler words

```python
import spacy

# Load SpaCy model
nlp = spacy.load('en_core_web_lg')

# Define the custom list of filler words
filler_words = set([
    'like', 'uh', 'um', 'ummm', 'hehehe', 'hahaha', 'you know', 'i mean', 'actually', 'literally', 
    'kinda', 'sort of', 'just', 'okay', 'so', 'well', 'right', 'anyway', 'uhm', 'erm', 'totally', 
    'seriously', 'honestly', 'pretty much', 'maybe', 'perhaps', 'really', 'sorta', 'yeah', 'yup', 
    'nope', 'oh', 'ah', 'alright', 'yknow', 'gotcha', 'look', 'obviously', 'clearly', 'apparently', 
    'definitely', 'absolutely', 'quite', 'nearly', 'only', 'sure', 'stuff', 'things', 'whatever', 
    'whenever', 'whoever', 'wherever', 'whatchamacallit', 'whatcha', 'yada', 'ok', 'kind of', 'anyhow', 
    'somewhat', 'around', 'etc', 'and so on', 'and so forth', 'seriously', 'totally', 'thus', 
    'therefore', 'at the end of the day', 'to be honest', 'believe me', 'by the way', 'furthermore', 
    'in fact', 'in my opinion', 'like I said', 'in a nutshell', 'in short', 'basically', 'on the other hand', 
    'as a matter of fact', 'incidentally', 'kindof', 'for sure', 'it was', 'sorta', 'kinda', 'and all','though',
    'certainly','instead', 'especially','mostly','although','particularly','either','back','given','mainly',
])

# Function to remove filler words based on the custom list
def remove_filler_words(text):
    # Process the text using SpaCy
    doc = nlp(text)
    # Keep tokens that are not in the filler words list
    tokens = [token.text for token in doc if token.text.lower() not in filler_words]
    return ' '.join(tokens)

# Apply the filler word removal function based on the custom filler words list
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_filler_words)
```

## Removing Numbers and Measurements

```python
import re
import inflect

# Initialize the inflect engine
p = inflect.engine()

def remove_numbers_and_measurements(text):
    # Remove numbers and measurements (e.g., "100kg", "5m", etc.)
    text = re.sub(r'\b\d+\w*\b', '', text)
    
    # Tokenize and filter out words representing numbers
    text = ' '.join([word for word in text.split() if not (p.number_to_words(word, wantlist=False).isdigit() or p.ordinal(word).isdigit())])
    
    return text

# Apply the function to remove numbers, measurements, and words representing numbers
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_numbers_and_measurements)
for i, review in enumerate(movie_reviews_df['review'].head(3)):
    print(f"Review {i + 1}: {review}")
```

## Removal of Hashtags, URLS and Emojis

```python
import re

# Function to remove URLs from text
def remove_urls(text):
    # Regex pattern to match URLs
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    # Substitute URLs with an empty string
    text = url_pattern.sub(r'', text)
    return text

# Function to remove mentions from text
def remove_mentions(text):
    # Regex pattern to match mentions (e.g., @username)
    mention_pattern = re.compile(r'@\w+')
    # Substitute mentions with an empty string
    return mention_pattern.sub(r'', text)

# Example usage on a review column
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_urls)
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_mentions)

# Optional: Print 3 sample reviews to verify
for i, review in enumerate(movie_reviews_df['review'].head(3)):
    print(f"Review {i + 1}: {review}")
```

Explanation:
Remove URLs:

The remove_urls function uses a regular expression to identify URLs and replaces them with an empty string.
Remove Mentions:

The remove_mentions function defines a regex pattern @\w+, which matches any mention that starts with @ followed by word characters (letters, digits, or underscores). It then substitutes these mentions with an empty string.
Applying Functions:

The remove_urls function is applied first, followed by the remove_mentions function to the review column of your DataFrame.
Verification:

Finally, a loop prints out the first three cleaned reviews to confirm that both URLs and mentions have been successfully removed.
This code will help ensure your text data is free from both URLs and user mentions, making it cleaner for further analysis.

## Removing Non-Dictionary Words from Movie Reviews

```python
import nltk
nltk.download('words')
```

```python
from nltk.corpus import words

# Get the list of valid English words
english_words = set(words.words())

# Function to remove words not found in the English dictionary
def remove_non_dictionary_words(text):
    tokens = text.split()  # Tokenize the text
    valid_words = [word for word in tokens if word.lower() in english_words]  # Keep only dictionary words
    return ' '.join(valid_words)

# Apply the function to your reviews
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_non_dictionary_words)

# Print 10 sample reviews to verify
for i, review in enumerate(movie_reviews_df['review'].head(3)):
    print(f"Review {i + 1}: {review}")
```

Loading English Words:

english_words = set(words.words()) loads a set of valid English words from the NLTK library’s corpus. Using a set ensures that the word lookups are efficient.
Function Purpose:

remove_non_dictionary_words(text) is a function that filters out any tokens (words) that are not found in the list of English words. It does this by tokenizing the input text, checking each word against english_words, and keeping only the words that match.
Applying the Function:

Applying this function to the review column of movie_reviews_df means that each review will be processed to remove words that aren’t recognized as standard English, such as misspelled words, slang, or foreign words.
Sample Verification:

Displaying a few sample reviews allows you to quickly check the effectiveness of this filtering step.
Importance:

This step is crucial for preprocessing text data in NLP, especially when working with applications like sentiment analysis or topic modeling, where it’s useful to have standardized, meaningful words.
Removing non-dictionary words improves the quality of the data by reducing noise, ensuring that the analysis focuses on legitimate English terms and reducing the impact of errors, typos, or irrelevant language on the results.

```python
import nltk
from nltk.corpus import words
import re

# Download the 'words' corpus if not already downloaded
nltk.download('words')

# Get the list of valid English words
english_words = set(words.words())

# List of common significant short words that should be retained
significant_short_words = {"not", "and", "the", "is", "to", "in", "of", "a", "it", "on"}

# Function to remove non-dictionary words and short words (1, 2, or 3 letters)
def remove_non_dictionary_and_short_words(text):
    tokens = text.split()  # Tokenize the text
    
    valid_words = [
        word for word in tokens 
        if (len(word) > 3 and word.lower() in english_words) or word.lower() in significant_short_words
    ]
    
    # Optionally, use regex to remove unwanted short words (1-3 characters) that are not in significant_short_words
    valid_words = [word for word in valid_words if not re.match(r'^\w{1,3}$', word)]
    
    return ' '.join(valid_words)

# Apply the function to your reviews
movie_reviews_df['review'] = movie_reviews_df['review'].apply(remove_non_dictionary_and_short_words)

# Print 10 sample reviews to verify
for i, review in enumerate(movie_reviews_df['review'].head(3)):
    print(f"Review {i + 1}: {review}")
```

This code is designed to clean movie reviews for natural language processing tasks by removing irrelevant or non-meaningful words. By eliminating short words and ensuring that only valid English words longer than three letters are retained, it enhances the quality of the text data. The final verification step allows for a quick inspection of the processed reviews, ensuring the filtering was executed correctly and that the remaining content is meaningful for subsequent analysis or modeling.

```python
import pandas as pd

# List of common negation words
negation_words = [
    "not", "never", "no", "none", "nobody", "nothing", "nowhere",
    "neither", "nor", "without", "hardly", "scarcely", "barely",
    "cannot", "can't", "could not", "couldn't", "is not", "isn't",
    "are not", "aren't", "was not", "wasn't", "were not", "weren't",
    "has not", "hasn't", "have not", "haven't", "had not", "hadn't",
    "should not", "shouldn't", "would not", "wouldn't", "might not",
    "mightn't", "must not", "mustn't", "don't", "do not", "doesn't",
    "does not", "won't", "will not", "ain't", "lacking", "no one",
    "never ever", "not a single"
]

# Function to handle negation in reviews
def handle_negation(words):
    modified_words = []  # List to hold modified words
    
    i = 0
    while i < len(words):
        if words[i].lower() in negation_words:  # Check for negation words
            if i + 1 < len(words):  # If there's a following word
                # Prepend "not_" to the following word
                modified_words.append(f"not_{words[i + 1]}")
                i += 2  # Skip the next word
            else:
                modified_words.append(words[i])  # Add negation word if it's the last word
                i += 1
        else:
            modified_words.append(words[i])  # Just add the word
            i += 1
    
    return modified_words  # Return the modified words as a list

# Apply the function to handle negation in your reviews
movie_reviews_df['review'] = movie_reviews_df['review'].apply(handle_negation)

# Print the modified reviews to verify the changes
for i, review in enumerate(movie_reviews_df['review'].head(3)):
    print(f"Review {i + 1}: {review}")
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def create_comparison_wordcloud(df):
    # Create figure with two subplots side by side
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(24, 12), facecolor='None')

    # Process positive reviews
    positive_reviews = df[df['sentiment'] == 'positive']['review']
    positive_text = ' '.join([' '.join(words) if isinstance(words, list) else words for words in positive_reviews])

    # Process negative reviews
    negative_reviews = df[df['sentiment'] == 'negative']['review']
    negative_text = ' '.join([' '.join(words) if isinstance(words, list) else words for words in negative_reviews])

    # Generate word cloud for positive reviews
    wordcloud_positive = WordCloud(
        max_words=500,
        width=800,
        height=800,
        background_color='white',
        colormap='YlGn'  # Yellow-Green colormap for positive sentiment
    ).generate(positive_text)

    # Generate word cloud for negative reviews
    wordcloud_negative = WordCloud(
        max_words=500,
        width=800,
        height=800,
        background_color='white',
        colormap='RdPu'  # Red-Purple colormap for negative sentiment
    ).generate(negative_text)

    # Plot positive reviews
    ax1.imshow(wordcloud_positive, interpolation='bilinear')
    ax1.axis('off')
    ax1.set_title('Most Frequent Words in Positive Reviews', fontsize=19, pad=20)

    # Plot negative reviews
    ax2.imshow(wordcloud_negative, interpolation='bilinear')
    ax2.axis('off')
    ax2.set_title('Most Frequent Words in Negative Reviews', fontsize=19, pad=20)

    # Add main title
    plt.suptitle('Comparison of Word Frequencies in Movie Reviews', fontsize=22, y=1.02)

    # Adjust layout to prevent overlap
    plt.tight_layout()

    return fig


fig = create_comparison_wordcloud(movie_reviews_df)
plt.show()
```

The goal of a machine learning model is to get better at understanding data over time by training on more data and finding patterns.

Input Data: First, you provide the model with data. This data could be numbers, text, images—anything.
Training: The model goes through a “training” process, where it studies the data and tries to recognize patterns. For example, if you train a model to recognize positive or negative movie reviews, you’d give it examples of both.

Clustering Models: These group similar data together. An example would be grouping customers into different segments based on their shopping behavior.
Recommendation Models: These suggest items to users. For example, Netflix recommending shows based on what you’ve previously watched.

##  Logistic Regression model for sentiment analysis by incorporating the ROC (Receiver Operating Characteristic) curve.

```python
from sklearn.metrics import roc_curve, auc

# Generate ROC curve data
y_prob_log_reg = log_reg.predict_proba(X_test_tfidf)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob_log_reg)
roc_auc = auc(fpr, tpr)

# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label='Logistic Regression (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Logistic Regression')
plt.legend(loc='lower right')
plt.show()

print(f"Logistic Regression Accuracy: {log_reg.score(X_test_tfidf, y_test):.2f}")
```

### ROC Curve Generation
The code first imports the necessary functions for generating the ROC curve and calculating the area under the curve (AUC) from the sklearn.metrics module.
Using the predict_proba method of the logistic regression model, it generates predicted probabilities of the positive class (i.e., sentiment labeled as 'positive'). This provides a probability score for each instance in the test set, enabling a more nuanced evaluation compared to simple predictions.

### Calculating FPR and TPR
The roc_curve function computes the False Positive Rate (FPR) and True Positive Rate (TPR) at various threshold settings. FPR indicates the proportion of actual negatives that were incorrectly classified as positives, while TPR (also known as recall) shows the proportion of actual positives correctly identified. This calculation helps visualize the trade-off between sensitivity and specificity at different thresholds.

### Plotting the ROC Curve
A plot is generated to display the ROC curve, which plots the TPR against the FPR. The curve provides a graphical representation of the model's performance across various threshold settings. The diagonal dashed line represents a random classifier, serving as a baseline to evaluate the model’s performance.
The area under the ROC curve (AUC) is calculated using the auc function, giving a single metric to quantify the overall ability of the model to discriminate between the positive and negative classes. An AUC of 1 indicates perfect discrimination, while an AUC of 0.5 indicates no discrimination (equivalent to random guessing).

### Visualizing and Reporting Accuracy
Finally, the ROC curve is displayed, showing the performance of the logistic regression model. Additionally, the accuracy of the model is printed again, reinforcing how well it performs overall in classifying sentiments.

### Significance of the Approach

Incorporating the ROC curve in model evaluation is important because:

It allows you to assess the model's performance at various decision thresholds, giving insight into how it might behave in different scenarios.
The AUC provides a clear, concise metric for evaluating model performance independent of the classification threshold, making it easier to compare with other models.
The ROC curve helps visualize the balance between sensitivity and specificity, which is particularly useful in imbalanced datasets where one class might dominate.
Overall, the ROC curve and AUC metrics offer deeper insights into the effectiveness of your Logistic Regression model for sentiment analysis, enabling more informed decisions on model selection and further improvements.
